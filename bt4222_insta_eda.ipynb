```python
import pandas as pd
captions = pd.read_csv('captions_2610.csv')


# captions['captions'] = [x.strip('()').split(',') for x in captions['captions']]
captions['captions'] = captions['captions'].str.strip('[]').str.split(',')
captions["emoji"] = ""
captions["mentions"] = ""
captions["hashtags"] = ""
captions["url"] = ""
captions = captions.explode('captions').reset_index()
captions['captions'][4]
captions['clean_captions'] = ''
captions
```

```python
import re
import pandas as pd
for word in range(len(captions)):
    cap = captions["captions"][word]
    if type(cap) == float:
        cap = ""
    word_list = cap.split(' ')
    word_list2 = [re.sub('[0-9]', ' ', i) for i in word_list]
    word_list3 = [re.sub("\d*\.\d+", " ", i) for i in word_list2]
    word_list4 = [str(i).replace('\n', ' ') for i in word_list3]
    word_list5 = [str(i).replace('\\n\\n', ' ') for i in word_list4]
    word_list6 = [str(i).replace('\\n', ' ') for i in word_list5]
    regex = "#(\w+)"
    hashtag_list = re.findall(regex, cap)
    hashtag_list_lower = [item.lower() for item in hashtag_list]
    new_caption = " ".join(word_list6)
    ls = new_caption.split(' ')
    for i in ls.copy():
        if i.startswith('#'):
            ls.remove(i) 
    new_caption = " ".join(ls)
    captions["clean_captions"][word] = new_caption
    captions["hashtags"][word] = hashtag_list
captions[:200]
```

```python
#take out mentions
for word in range(len(captions)):
    cap = captions["clean_captions"][word]
    word_list = cap.split(' ')
    regex = "@(\w+)"
    mention_list = re.findall(regex, cap)
    mention_list_lower = [item.lower() for item in mention_list]
    new_caption = " ".join(word_list)
    ls = new_caption.split(' ')
    for i in ls.copy():
        if i.startswith('@'):
            ls.remove(i) 
    new_caption = " ".join(ls)
    captions["clean_captions"][word] = new_caption
    captions["mentions"][word] = mention_list
captions[:200]
```

```python
#take out url
for word in range(len(captions)):
    cap = captions["clean_captions"][word]
    word_list = cap.split(' ')
    regex = '(?:(?:https?|ftp):\/\/)?[\w/\-?=%.]+\.[\w/\-&?=%.]+'
    url_list = re.findall(regex, cap)
    url_list_lower = [item.lower() for item in url_list]
    new_caption = " ".join(word_list)
    ls = new_caption.split(' ')
    for i in ls.copy():
        if i in url_list_lower:
            ls.remove(i) 
    new_caption = " ".join(ls)
    captions["clean_captions"][word] = new_caption
    captions["url"][word] = url_list
captions[:200]
```

```python
#take out emojis
import demoji
import emoji

def remove_emoji(string):
    return emoji.get_emoji_regexp().sub(u'', string)

for word in range(len(captions)):
    cap = captions["clean_captions"][word]
    word_list = cap.split(' ')
    emoji_list = demoji.findall(cap)
    new_caption = " ".join(word_list)
    ls = new_caption.split(' ')
    ls = [demoji.replace(i, "") for i in ls]
    new_caption = " ".join(ls)
    captions["clean_captions"][word] = new_caption
    captions["emoji"][word] = emoji_list
captions[:200]
```

```python
import textblob
from textblob import Word
captions['incorrect_spell_count'] = ''
captions['incorrect_spell_words'] = ''
for word in range(len(captions)):
    cap = captions["clean_captions"][word]
    word_list = cap.split(' ')
    correct_words = []
    incorrect_words = []
    for i in word_list:
        if Word(i).spellcheck()[0][1] > 0.7:
            correct_words.append(i)
        else:
            incorrect_words.append(i)
    new_caption = " ".join(correct_words)
    captions["clean_captions"][word] = new_caption
    captions['incorrect_spell_words'][word] = incorrect_words
    captions['incorrect_spell_count'][word] = len(incorrect_words)
captions[:200]
    
    
    
```

```python
#grammar check
import language_tool_python
tool = language_tool_python.LanguageTool('en-US')
captions['no_of_grammar_errors'] = ''
captions['%_of_improper_grammar'] = ''
for word in range(len(captions)):
    cap = captions["clean_captions"][word]
    word_list = cap.split(' ')
    captions['no_of_grammar_errors'][word] = len(tool.check(cap))
    captions['%_of_improper_grammar'][word] =len(tool.check(cap))/len(word_list)
captions[:200]
```
